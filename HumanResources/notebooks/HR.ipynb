```python
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x == 'Y' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
employee_df.head()
```

```python
# check missing data

sns.heatmap(employee_df.isnull(), yticklabels=False, cbar=False, cmap='Blues');
```

```python
employee_df.hist(bins=30, figsize=(20, 20), color='b');
```

```python
# drop irrelevant data
employee_df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
employee_df.info()
```

```python
left_df = employee_df[employee_df['Attrition'] == 1]
stayed_df = employee_df[employee_df['Attrition'] == 0]
```

```python
print("Number of employees who left company:", len(left_df))
print("% of employees who left company:", 1.0 * len(left_df) / len(employee_df) * 100)


print("Numbere of employees who stayed:", len(stayed_df))
print("% of employees who stayed:", 1.0 * len(stayed_df) / len(employee_df) * 100)
```

```python
correlations = employee_df.corr()
f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(correlations, annot=True);
```

```python
plt.figure(figsize=[25, 12])
sns.countplot(x='Age', hue='Attrition', data=employee_df);
```

```python
plt.figure(figsize=[20,20])

plt.subplot(411)
sns.countplot(x='JobRole', hue='Attrition', data=employee_df)

plt.subplot(412)
sns.countplot(x='MaritalStatus', hue='Attrition', data=employee_df)

plt.subplot(413)
sns.countplot(x='JobInvolvement', hue='Attrition', data=employee_df)

plt.subplot(414)
sns.countplot(x='JobLevel', hue='Attrition', data=employee_df);
```

```python
sns.kdeplot(left_df['DistanceFromHome'], label='Employees who left:', shade=True, color='orange')
sns.kdeplot(stayed_df['DistanceFromHome'], label='Employees who stayed', shade=True, color='blue')

plt.xlabel('Distance From Home');
```

```python
plt.figure(figsize=(12,7))

sns.kdeplot(left_df['YearsWithCurrManager'], label='Employees who left', shade=True, color='orange')
sns.kdeplot(stayed_df['YearsWithCurrManager'], label='Employees who Stayed', shade=True, color='blue')

plt.xlabel('Years With Current Manager');
```

```python
plt.figure(figsize=(12,7))

sns.kdeplot(left_df['TotalWorkingYears'], shade=True, label='Employees who left', color ='orange')
sns.kdeplot(stayed_df['TotalWorkingYears'], shade=True, label='Employees who Stayed', color ='blue')

plt.xlabel('Total Working Years');
```
